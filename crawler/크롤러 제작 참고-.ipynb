```python
from selenium import webdriver
from bs4 import BeautifulSoup as bs
from selenium.webdriver.common.keys import Keys
import time
import requests
import shutil
import pandas as pd
from urllib.request import urlopen
```

```python
dic = {'img_name' : [], 'img_src' : [], 'location' : [], 'comment' : [], 'hash_0' : [], 'hash_1' : [],'hash_2' : [],'hash_3' : [],'hash_4' : [],'hash_5' : [],'hash_6' : [],'hash_7' : [],'hash_8' : [],'hash_9' : [],'hash_10' : [],'hash_11' : [],'hash_12' : [],'hash_13' : [],'hash_14' : [],'hash_15' : [],'hash_16' : [],'hash_17' : [],'hash_18' : [],'hash_19' : [],'hash_all' : []}
```

```python
#창을 두개 띄워서 작업하고 둘다 로그인 되어 있어야 함

driver.get("https://www.instagram.com/accounts/login/")
driver_2.get("https://www.instagram.com/accounts/login/")
time.sleep(2);
```

```python
#로그인
driver.find_element_by_name('username').send_keys('본인 ID')
driver.find_element_by_name('password').send_keys('본인 PW')
driver_2.find_element_by_name('username').send_keys('본인 ID')
driver_2.find_element_by_name('password').send_keys('본인 PW')

time.sleep(2)
driver.find_element_by_xpath('//*[@id="loginForm"]/div/div[3]/button').click()
driver_2.find_element_by_xpath('//*[@id="loginForm"]/div/div[3]/button').click()
```

```python
#돋보기 표시 클릭
driver.find_element_by_xpath('//*[@id="react-root"]/section/nav/div[2]/div/div/div[3]/div/div[3]/a').click()
time.sleep(2)
```

```python

image_num = 0

for j in range(1,3) :
    for i in range(1,4) :
        #url_crawl = driver.find_element_by_xpath('//*[@id="react-root"]/section/main/article/div[1]/div/div/div[1]/div[%d]/a'%i)
        url_crawl = driver.find_element_by_xpath('//*[@id="react-root"]/section/main/div/div[1]/div/div[%d]/div[%d]/div/a'%(j,i))
        url = url_crawl.get_attribute('href')

        print(url)
        driver_2.get(url)
        time.sleep(2)
        location = driver_2.find_elements_by_xpath('//*[@id="react-root"]/section/main/div/div[1]/article/header/div[2]/div[2]/div[2]/a')
        if len(location)==0 :
            location = ""
        else : location = location[0].text


        comment = driver_2.find_elements_by_xpath('//*[@id="react-root"]/section/main/div/div[1]/article/div[3]/div[1]/ul/div/li/div/div/div[2]/span')
        if len(comment)==0 :
            comment = ""
        else : comment = comment[0].text


        #답글보기가 되어있는 모든 버튼 수집 후 클릭
        button_list = driver_2.find_elements_by_class_name('_61Di1')
        for button in button_list :
            button.click()

        time.sleep(1)
        #본문 글, 댓글 , 답글 상관없이 해시태그가 있는 모든 a 태그를 수집
        cur_page = driver_2.page_source
        hash_soup = bs(cur_page,'lxml')
        hash_list = hash_soup.find_all('a', class_='xil3i')


        #해시태그가 30개가 넘으면 다음 이미지로 넘어감
        if len(hash_list) > 20 : 
            continue
        else : image_num += 1


        img_src = driver_2.find_element_by_class_name('FFVAD')
        img_src = img_src.get_attribute('src')
        with urlopen(img_src) as f:
            with open('./img/' + 'img_' + str(image_num) + '.jpg', 'wb') as h:
                img = f.read()
                h.write(img)

        dic['img_name'].append('img'+'_'+str(image_num)+'.jpg')
        dic['img_src'].append(img_src)
        dic['location'].append(location)
        dic['comment'].append(comment)

        hash_all = ""
        for i,hash_ in enumerate(hash_list) :
            dic['hash_%d'%i].append(hash_.text)
            hash_all += hash_.text+" "
            print(hash_.text)

        for i in range(len(hash_list),20) :
            dic['hash_%d'%i].append("")

        dic['hash_all'].append(hash_all)
```

```python
df = pd.DataFrame(dic)
df
```

```python
#Daraframe 을 csv 로 만들기 
df.to_csv("data.csv", index=True, encoding='utf-8-sig')
```

```python

image_num = 0

for j in range(1,3) :
    for i in range(1,4) :
        #url_crawl = driver.find_element_by_xpath('//*[@id="react-root"]/section/main/article/div[1]/div/div/div[1]/div[%d]/a'%i)
        url_crawl = driver.find_element_by_xpath('//*[@id="react-root"]/section/main/div/div[1]/div/div[%d]/div[%d]/div/a'%(j,i))
        url = url_crawl.get_attribute('href')

        print(url)
        driver_2.get(url)
        time.sleep(2)
        location = driver_2.find_elements_by_xpath('//*[@id="react-root"]/section/main/div/div[1]/article/header/div[2]/div[2]/div[2]/a')
        if len(location)==0 :
            location = ""
        else : location = location[0].text


        comment = driver_2.find_elements_by_xpath('//*[@id="react-root"]/section/main/div/div[1]/article/div[3]/div[1]/ul/div/li/div/div/div[2]/span')
        if len(comment)==0 :
            comment = ""
        else : comment = comment[0].text


        #ë‹µê¸€ë³´ê¸°ê°€ ë˜ì–´ìˆëŠ” ëª¨ë“  ë²„íŠ¼ ìˆ˜ì§‘ í›„ í´ë¦­
        button_list = driver_2.find_elements_by_class_name('_61Di1')
        for button in button_list :
            button.click()

        time.sleep(1)
        #ë³¸ë¬¸ ê¸€, ëŒ“ê¸€ , ë‹µê¸€ ìƒê´€ì—†ì´ í•´ì‹œíƒœê·¸ê°€ ìˆëŠ” ëª¨ë“  a íƒœê·¸ë¥¼ ìˆ˜ì§‘
        cur_page = driver_2.page_source
        hash_soup = bs(cur_page,'lxml')
        hash_list = hash_soup.find_all('a', class_='xil3i')


        #í•´ì‹œíƒœê·¸ê°€ 30ê°œê°€ ë„˜ìœ¼ë©´ ë‹¤ìŒ ì´ë¯¸ì§€ë¡œ ë„˜ì–´ê°
        if len(hash_list) > 20 : 
            continue
        else : image_num += 1


        img_src = driver_2.find_element_by_class_name('FFVAD')
        img_src = img_src.get_attribute('src')
        with urlopen(img_src) as f:
            with open('./img/' + 'img_' + str(image_num) + '.jpg', 'wb') as h:
                img = f.read()
                h.write(img)

        dic['img_name'].append('img'+'_'+str(image_num)+'.jpg')
        dic['img_src'].append(img_src)
        dic['location'].append(location)
        dic['comment'].append(comment)

        hash_all = ""
        for i,hash_ in enumerate(hash_list) :
            dic['hash_%d'%i].append(hash_.text)
            hash_all += hash_.text+" "
            print(hash_.text)

        for i in range(len(hash_list),20) :
            dic['hash_%d'%i].append("")

        dic['hash_all'].append(hash_all)
```
